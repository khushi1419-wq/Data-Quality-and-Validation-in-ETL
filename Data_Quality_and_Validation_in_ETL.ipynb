**Question 1 : Define Data Quality in the context of ETL pipelines. Why is it more than just data cleaning?**

Answer:-

Data Quality in ETL pipelines refers to the accuracy, completeness, consistency, reliability, and validity of data as it moves through the Extract, Transform, and Load process. High-quality data ensures that business reports, analysis, and decisions are correct and trustworthy.

Data quality is more than just data cleaning because it involves the overall management and monitoring of data throughout the entire ETL process. Data cleaning mainly focuses on fixing errors, removing duplicates, and handling missing values, while data quality also includes data validation, standardization, consistency checks, data governance, and continuous monitoring.

Therefore, data quality ensures that data is not only clean but also useful, accurate, and reliable for business operations and decision-making.


**Question 2 : Explain why poor data quality leads to misleading dashboards and incorrect decisions.**

Answer:-

Poor data quality leads to misleading dashboards and incorrect decisions because inaccurate, incomplete, or inconsistent data produces unreliable results. If the data used in reports contains errors, missing values, or duplicates, the dashboard may show wrong trends, sales figures, or business performance.

**For example**, incorrect sales data can make management believe that a product is performing well when it is actually underperforming. Similarly, duplicate customer records may inflate customer counts and affect business analysis.

**Question 3 : What is duplicate data? Explain three causes in ETL pipelines.**

Answer:-

Duplicate data refers to repeated records or entries that appear more than once in a dataset. Duplicate records reduce data accuracy and can lead to incorrect analysis and reporting.

Three common causes of duplicate data in ETL pipelines are:

1. Multiple Data Sources – When data is collected from different systems, the same record may exist in more than one source and get loaded multiple times.

2. Lack of Unique Identifiers – If datasets do not have proper primary keys or unique IDs, the ETL process may fail to identify duplicate records.

3. Repeated ETL Loads – Sometimes ETL jobs run multiple times due to system errors or scheduling issues, causing the same data to be inserted again into the target database.

**Question 4 : Differentiate between exact, partial, and fuzzy duplicates.**

Answer:-

1. Exact Duplicates

Exact duplicates are records that are completely identical in all fields and values.

Example:
Two rows with the same customer name, phone number, and email.

2. Partial Duplicates

Partial duplicates are records where some fields are the same, but other fields are different or missing.

Example:
One record has the full address while another has only the city and phone number.

3. Fuzzy Duplicates

Fuzzy duplicates are records that are similar but not exactly the same due to spelling mistakes, formatting differences, or typing errors.

Example:
“Rahul Sharma” and “Rahul S.” or “Infosys Ltd” and “InfoSys Limited”.



**Question 5 : Why should data validation be performed during transformation rather than after loading?**

Answer:-

Data validation should be performed during the transformation stage because it helps detect and correct errors before data is loaded into the target system. This ensures that only accurate, complete, and properly formatted data enters the database or data warehouse.

If validation is done after loading, incorrect or invalid data may already affect reports, dashboards, and business analysis. Fixing errors later can also be more time-consuming and costly.

**Question 6 : Explain how business rules help in validating data accuracy. Give an example.**

Answer:-

Business rules are predefined conditions or standards used to check whether data is correct, valid, and meaningful according to business requirements. These rules help identify errors, inconsistencies, and invalid values during the ETL process.

For example, a business rule may state that a customer’s age cannot be negative and must be between 18 and 60 for a loan application. If the entered age is 75 or -5, the data will be rejected or flagged as invalid.
